```python
import datetime

# 取得當前日期
today = datetime.date.today()

# 設定每個月的早上11點
start_times = []
end_times = []

# 迴圈生成每個月的所有天數
for year in range(today.year, today.year + 1):  # 若要包含明年，可修改成 (today.year, today.year + 2)
    for month in range(today.month, 13):  # 若要包含全年，可修改成 range(1, 13)
        # 取得該月份的最後一天
        last_day_of_month = datetime.date(year, month % 12 + 1, 1) - datetime.timedelta(days=1)

        # 迴圈設定每一天的早上11點
        for day in range(1, last_day_of_month.day + 1):
            date = datetime.date(year, month, day)
            start = datetime.datetime.combine(date, datetime.time(11, 0))
            end = datetime.datetime.combine(date, datetime.time(11, 0))

            # 添加到列表中
            start_times.append(start)
            end_times.append(end)

# 確認生成的時間列表
for i in range(len(start_times)):
    print(f"第 {i+1} 個時間段:")
    print("開始時間:", start_times[i])
    print("結束時間:", end_times[i])
```

```python
import datetime
import calendar
import pandas as pd

def get_month_dates(year, month):
    num_days = calendar.monthrange(year, month)[1]
    dates = []
    for day in range(1, num_days + 1):
        date = datetime.date(year, month, day)
        dates.append(date)
    return dates


def replace_dict_with_zero(value):
    if isinstance(value, dict):
        return 0
    return value
```
